```python


record_count = 0

# stores the mapping from a person to the people they follow
follows = {}

unique_ids = set({})
total_follows_count = 0
min = 100000000
max = -1

# tracks the number of people who follow nobody
zero_follows_records = 0

for entry in train_data:

  data = entry.strip().split('\t')
  # convert string into an array of ids
  ids = [int(id_string) for id_string in data]

  # add all the entries to the SET to find the unique ids in the training data
  unique_ids.update(ids)

  # extract the user in question
  user_id = ids[0]

  # extract the people who the user follows, by removing the user-id from the ids
  ids = ids[1:]

  # get the count of people who the user follows
  follows_count = len(ids)

  # update total records count
  record_count = record_count + 1

  # maps user to the people they follow
  follows[user_id] = ids
  total_follows_count = total_follows_count + follows_count

  # if the people a user follows is zero, update the zero_follow record count
  if follows_count == 0:
    zero_follows_records += 1

  # calculate min and max followed count
  if follows_count > max:
    max = follows_count
  if follows_count < min:
    min = follows_count


print("Train data statistics")
print('Total records         : ', record_count)
print('Total unique ids      : ', len(unique_ids))
print('Total linkages        : ', total_follows_count)
print('Max no. of people followed by a single person  : ', max)
print('Min no. of people followed by a single person  : ', min)
print('No. of users who follow nobody : ', zero_follows_records)

```

```python
# Create a reverse mapping from a user to the people who follow them

# stores the mapping from a person to the people who follows them
followers = {}


# random variable(no of people follow = input)
follows_count_frequency = {}

for follower in follows:

  # get the number of people a person follows
  follows_count = len(follows[follower])

  # calculate a distribution of no of followed persons for different people
  # KEY - #followed people    VALUE - #followers who have KEY(#followed) number of followed people
  if follows_count in follows_count_frequency:
    follows_count_frequency[follows_count] = follows_count_frequency[follows_count] + 1
  else:
    follows_count_frequency[follows_count] = 1


  # get all the people a follower, follows
  followed_persons = follows[follower]
  # create a reverse mapping
  for followed_person in followed_persons:
    if followed_person in followers:
      followers[followed_person].append(follower)
    else:
      followers[followed_person] = [follower]



print(follows_count_frequency)
```

```python
import logging
from neo4j import GraphDatabase
from neo4j.exceptions import ServiceUnavailable

class App:

    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        # Don't forget to close the driver connection when you are finished with it
        self.driver.close()

    def create_follow_relation(self, person1_name, person2_name):
        with self.driver.session() as session:
            # Write transactions allow the driver to handle retries and transient errors
            result = session.write_transaction(
                self._create_and_return_follow_relation, person1_name, person2_name)

    @staticmethod
    def _create_and_return_follow_relation(tx, person1_name, person2_name):

        # To learn more about the Cypher syntax,
        # see https://neo4j.com/docs/cypher-manual/current/

        # The Reference Card is also a good resource for keywords,
        # see https://neo4j.com/docs/cypher-refcard/current/

        query = (
            "MERGE (p1:Person { name: $person1_name }) "
            "MERGE (p2:Person { name: $person2_name }) "
            "MERGE (p1)-[:FOLLOWS]->(p2) "
            "RETURN p1, p2"
        )
        result = tx.run(query, person1_name=person1_name, person2_name=person2_name)
        try:
            return [{"p1": record["p1"]["name"], "p2": record["p2"]["name"]}
                    for record in result]
        # Capture any errors along with the query and data for traceability
        except ServiceUnavailable as exception:
            logging.error("{query} raised an error: \n {exception}".format(
                query=query, exception=exception))
            raise

    def find_person(self, person_name):
        with self.driver.session() as session:
            result = session.read_transaction(self._find_and_return_person, person_name)
            for record in result:
                print("Found person: {record}".format(record=record))
                return record

    @staticmethod
    def _find_and_return_person(tx, person_name):
        query = (
            "MATCH (p:Person) "
            "WHERE p.name = $person_name "
            "RETURN p.name AS name"
        )
        result = tx.run(query, person_name=person_name)
        return [record["name"] for record in result]


# See https://neo4j.com/developer/aura-connect-driver/ for Aura specific connection URL.
# scheme = "neo4j"  # Connecting to Aura, use the "neo4j+s" URI scheme
# host_name = "example.com"
# port = 7687
# url = "{scheme}://{host_name}:{port}".format(scheme=scheme, host_name=host_name, port=port)
# user = "<Username for Neo4j database>"
# password = "<Password for Neo4j database>"
uri = "neo4j://localhost:7687"
driver = GraphDatabase.driver(uri, auth=("neo4j", "karun"))


app = App(uri, "neo4j", "karun")
app.create_follow_relation("Alice", "David")
app.create_follow_relation("Bill", "Alice")
a = app.find_person("Alice")


for follower in follows:
    # get all the people a follower, follows
    followed_persons = follows[follower]
    # create a follows relation
    for followed_person in followed_persons:
        app.create_follow_relation(follower, followed_person)

app.close()
```
